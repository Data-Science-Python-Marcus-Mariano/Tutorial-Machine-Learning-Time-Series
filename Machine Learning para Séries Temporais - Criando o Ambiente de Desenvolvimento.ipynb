## Gas Prices in Brazil


**Gas Prices in Brazil from 2004 to 2019: [Start here!](https://www.kaggle.com/matheusfreitag/gas-prices-in-brazil)**

```python
import pandas as pd
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt

%matplotlib inline

# Set quantity of max columns
pd.set_option('display.max_columns', 200)
```

```python
df = pd.read_csv('data/2004-2019.tsv', sep='\t') 
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['PRODUTO'].value_counts()
```

```python
diesel = df[df['PRODUTO'] == 'ÓLEO DIESEL'].copy()
```

```python
diesel.head()
```

```python
diesel['DATA INICIAL'] = pd.to_datetime(diesel['DATA INICIAL'])
diesel['DATA FINAL'] = pd.to_datetime(diesel['DATA FINAL'])
```

```python
(diesel['DATA FINAL'] - diesel['DATA INICIAL']).value_counts()
```

```python
diesel_train = diesel[diesel['DATA FINAL'] < '2011-01-01']
diesel_valid = diesel[diesel['DATA FINAL'] >= '2011-01-01']
diesel_train.shape, diesel_valid.shape
```

## Baseline

```python
y_train = diesel_train['PREÇO MÉDIO REVENDA']
y_valid = diesel_valid['PREÇO MÉDIO REVENDA']
```

```python
# function shift(1) mean price of week.
baseline_train = diesel_train.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].shift(1)
baseline_valid = diesel_valid.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].shift(1)
```

```python
# Compare baseline to mean price
diesel_train['baseline'] = baseline_train
```

```python
diesel_train[diesel_train['ESTADO'] == 'SAO PAULO'].head()
```

```python
from sklearn.metrics import mean_squared_log_error

# 
# Root Mean Square Erro log
print('RMSE:', np.sqrt(mean_squared_log_error(y_valid[baseline_valid.notnull()], 
                                              baseline_valid[baseline_valid.notnull()])))
                        
```
